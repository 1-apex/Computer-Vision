```python
import cv2 
from matplotlib import pyplot as plt 
```

```python
img = cv2.imread('Images/Expt 05/car.jpg',0) 
```

```python
histr = cv2.calcHist([img],[0],None,[256],[0,256]) 
```

```python
plt.plot(histr) 
plt.show()
```
